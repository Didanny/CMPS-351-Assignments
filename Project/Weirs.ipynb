```python
from pyeasyga import pyeasyga as ga
```

```python
data = [('pear', 50), ('apple', 35), ('banana', 40), ('kiwi', 60)]
```

```python
g = ga.GeneticAlgorithm(data)
```

```python
def fitness (individual, data):
    fitness = 0
    print('Individual: ')
    print(individual)
    if individual.count(1) == 2:
        for (selected, (fruit, profit)) in zip(individual, data):
            if selected:
                fitness += profit
    return fitness
```

```python
g.fitness_function = fitness
```

```python
g.run()
```

```python
print (g.best_individual())
```
